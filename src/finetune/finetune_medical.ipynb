```python
from unsloth import FastLanguageModel
import torch

# Thay đổi model_name
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "arcee-ai/Arcee-VyLinh",
    max_seq_length = 2048, # Hoặc 4096 nếu cần
    load_in_4bit = True,
)

# Cấu hình LoRA cho Llama 2
# Unsloth sẽ tự động tìm các module phù hợp
model = FastLanguageModel.get_peft_model(
    model,
    r = 16,
    lora_alpha = 16,
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj"],
    lora_dropout = 0,
    bias = "none",
    use_gradient_checkpointing = True,
)
model
```

### Setup train

```python
from datasets import load_dataset
import unicodedata

def normalize_text(s):
    return unicodedata.normalize("NFC", s.strip())

from datasets import load_dataset
raw = load_dataset("hungnm/vietnamese-medical-qa")
def prep(ex):
    return {
        "question": normalize_text(ex["question"]),
        "answer": normalize_text(ex["answer"])
    }
dataset = raw["train"].map(prep)

# Chia train/validation nếu dữ liệu chưa chia sẵn
dataset = dataset.train_test_split(test_size=0.1)
dataset
# def preprocess(sample):
#     prompt = f"Question: {sample['question']}\nAnswer:"
#     return {"input": prompt, "output": f" {sample['answer']}"}

# dataset = dataset.map(preprocess, remove_columns=dataset["train"].column_names)
```

```python
from transformers import TrainingArguments, EarlyStoppingCallback
from trl import SFTTrainer

def formatting_prompts_func(examples):
    questions = examples["question"]
    answers = examples["answer"]
    texts = []
    # System prompt không thay đổi
    system_prompt = "Bạn là trợ lý y tế chuyên nghiệp."

    for question, answer in zip(questions, answers):
        # Tạo prompt theo đúng định dạng Llama 2
        text = f"[INST] <<SYS>>\\n{system_prompt}\\n<</SYS>>\\n\\n{question} [/INST] {answer}" + tokenizer.eos_token
        texts.append(text)

    return { "text" : texts, }
dataset = dataset.map(formatting_prompts_func, batched = True,)
dataset['train'][0]
```

```python

training_args = TrainingArguments(
    per_device_train_batch_size = 4,
    gradient_accumulation_steps = 4,
    warmup_steps = 100,
    num_train_epochs = 3,

    learning_rate = 2e-4,

    bf16 = True,
    logging_steps = 25,
    output_dir = "finetuned_medical_qa_full",
    optim = "adamw_8bit",
    eval_strategy = "steps",
    eval_steps = 100,
    save_strategy = "steps",
    save_steps = 100,
    load_best_model_at_end = True,
    metric_for_best_model = "eval_loss",
    greater_is_better = False,
    save_total_limit = 2,
)

trainer = SFTTrainer(
    model=model,
    train_dataset=dataset["train"],
    eval_dataset=dataset["test"],
    tokenizer=tokenizer,
    args=training_args,
    dataset_text_field="text",
    callbacks=[EarlyStoppingCallback(early_stopping_patience=3)],
    max_seq_length=2048,
    report_to="none"
    # patience=3: nếu 3 lần eval liên tiếp không cải thiện thì dừng
)

trainer.train()
```

```python
from transformers import TextStreamer
prompt = (
    "<|system|>\nBạn là trợ lý y tế chuyên nghiệp.\n"
    "<|user|>\nBệnh sởi là bệnh gì?\n"
    "<|assistant|>\n"
)
input_ids = tokenizer(prompt, return_tensors="pt").to("cuda")
streamer = TextStreamer(
    tokenizer,
    skip_prompt=True,        # không in lại prompt
    skip_special_tokens=True # bỏ qua các token đặc biệt
)
outputs = model.generate(
    **input_ids,
    max_new_tokens=512,
    do_sample=True,
    temperature=0.7,
    top_k=50,
    top_p=0.9,
    repetition_penalty=1.2,
    no_repeat_ngram_size=2,
    eos_token_id=tokenizer.eos_token_id,
    streamer=streamer
)
# print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```

```python
# Save the fine-tuned model and tokenizer
model.save_pretrained("VyLinh_finetuned_medical_qa")
tokenizer.save_pretrained("Vy_Linh_finetuned_medical_qa")
```
